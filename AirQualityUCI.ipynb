You can see that NMHC_GT median value is 0, that's make posibble if NMHC_GT value contains many 0 values. so that's why i fill the missing value with mean value. But, in some cases like PT08_S4_NO2 and PT08_S5_O3 that have difference quite large between mean and median, possibility because both of this column have a lot of pretty high value. So, i choose to fill both of this attributes with median value.

```python
for i in data_air_quality.ix[data_air_quality['CO(GT)']==0].index:
    data_air_quality.loc[i, 'CO(GT)'] = data_air_quality['CO(GT)'].mean()
for i in data_air_quality.ix[data_air_quality['PT08.S1(CO)']==0].index:
    data_air_quality.loc[i, 'PT08.S1(CO)'] = data_air_quality['PT08.S1(CO)'].mean()
for i in data_air_quality.ix[data_air_quality['NMHC(GT)']==0].index:
    data_air_quality.loc[i, 'NMHC(GT)'] = data_air_quality['NMHC(GT)'].mean()
for i in data_air_quality.ix[data_air_quality['PT08.S2(NMHC)']==0].index:
    data_air_quality.loc[i, 'PT08.S2(NMHC)'] = data_air_quality['PT08.S2(NMHC)'].mean()
for i in data_air_quality.ix[data_air_quality['NOx(GT)']==0].index:
    data_air_quality.loc[i, 'NOx(GT)'] = data_air_quality['NOx(GT)'].median()
for i in data_air_quality.ix[data_air_quality['PT08.S3(NOx)']==0].index:
    data_air_quality.loc[i, 'PT08.S3(NOx)'] = data_air_quality['PT08.S3(NOx)'].mean()
for i in data_air_quality.ix[data_air_quality['NO2(GT)']==0].index:
    data_air_quality.loc[i, 'NO2(GT)'] = data_air_quality['NO2(GT)'].mean()
for i in data_air_quality.ix[data_air_quality['PT08.S4(NO2)']==0].index:
    data_air_quality.loc[i, 'PT08.S4(NO2)'] = data_air_quality['PT08.S4(NO2)'].median()
for i in data_air_quality.ix[data_air_quality['PT08.S5(O3)']==0].index:
    data_air_quality.loc[i, 'PT08.S5(O3)'] = data_air_quality['PT08.S5(O3)'].median()
for i in data_air_quality.ix[data_air_quality['T']==0].index:
    data_air_quality.loc[i, 'T'] = data_air_quality['T'].mean()
for i in data_air_quality.ix[data_air_quality['RH']==0].index:
    data_air_quality.loc[i, 'RH'] = data_air_quality['RH'].mean()
for i in data_air_quality.ix[data_air_quality['AH']==0].index:
    data_air_quality.loc[i, 'AH'] = data_air_quality['AH'].mean()
```

# Linear Regression

```python
from sklearn.preprocessing import StandardScaler         #import normalisation package
from sklearn.model_selection import train_test_split      #import train test split
from sklearn.linear_model import LinearRegression         #import linear regression package
from sklearn.metrics import mean_squared_error,mean_absolute_error   #import mean squared error and mean absolute error
```

```python
X=data_air_quality[col_].drop('RH',1)     #X-input features
y=data_air_quality['RH']  
```

```python
ss=StandardScaler()     #initiatilise
```

```python
X_std=ss.fit_transform(X)     #apply stardardisation
```

Train test split

```python
#split the data into train and test with test size and 30% and train size as 70%
X_train, X_test, y_train, y_test=train_test_split(X_std,y,test_size=0.3, random_state=42)
```

```python
print('Training data size:',X_train.shape)
print('Test data size:',X_test.shape)
```

```python
lr=LinearRegression()
lr_model=lr.fit(X_train,y_train)          #fit the linear model on train data
```

```python
print('Intercept:',lr_model.intercept_)
print('--------------------------------')
print('Slope:')
list(zip(X.columns.tolist(),lr_model.coef_))
```

Prediction

```python
y_pred=lr_model.predict(X_test)                      #predict using the model
rmse=np.sqrt(mean_squared_error(y_test,y_pred))      #calculate rmse
print('Baseline RMSE of model:',rmse)
```

 Conclusion of baseline linear regression model:
This means that we can predict RH using all the features together with RMSE as 6.01. Let us call it as baseline model.

```python
# write function to measure RMSE
def train_test_RMSE(feature):
    X=data_air_quality[feature]
    y=data_air_quality['RH']
    X_std_one=ss.fit_transform(X)
    X_trainR,X_testR,y_trainR,y_testR=train_test_split(X_std_one,y,test_size=0.3,random_state=42)
    lr_model_one=lr.fit(X_trainR,y_trainR)
    y_predR=lr_model_one.predict(X_testR)
    return np.sqrt(mean_squared_error(y_testR,y_predR))
```
